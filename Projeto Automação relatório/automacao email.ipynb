Para resolver isso, vamos foi utilizado o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time

pyautogui.PAUSE = 0.5

# Passo a passo do desafio
# Passo 1: Entrar no sistema da empresa (link do drive)
# abrir o Chrome
pyautogui.press("win")
pyautogui.write("chrome")
pyautogui.press("enter")
time.sleep(2)
pyautogui.click(x=875, y=502)


time.sleep(2)

# digita o link do sistema
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga" 
pyautogui.write(link)
# aperta enter e aguarda
pyautogui.press("enter")
time.sleep(3)
```

```python
# Passo 2: Navegar no sistema para encontrar a base de dados
pyautogui.click(x=394, y=515, clicks=2)
time.sleep(1)

# Passo 3: Exportar a base de dados (baixar o arquivo)
pyautogui.click(x=401, y=510) # seleciona o arquivo
pyautogui.click(x=1810, y=518) # clica nos 3 pontos
pyautogui.click(x=1547, y=610) # baixa o arquivo
time.sleep(3)
```

```python
import pandas

# Passo 4: Calcular os indicadores (faturamento e quantidade de produtos vendidos)
# abrir a base de dados
caminho = r"C:\Users\joaol\Downloads\Vendas - Dez.xlsx"     
tabela = pandas.read_excel(caminho)

# ver as informações da base de dados
display(tabela)

# somar o faturamento de todos os produtos = somar a coluna de Valor Final
faturamento = tabela["Valor Final"].sum()
# somar a quantidade de produtos = somar a coluna de quantidade
qtde_produtos = tabela["Quantidade"].sum()

print(faturamento)
print(qtde_produtos)
```

```python
import pyperclip

# Passo 5: Enviar as informações por e-mail
# abrir uma nova aba
pyautogui.hotkey("ctrl", "t")

# entrar no email
pyautogui.write("http://mail.google.com/")
pyautogui.press("enter")
time.sleep(8)

# clicar no botão escrever
pyautogui.click(x=188, y=222)
time.sleep(6)

# para quem enviar o email
pyautogui.write("leohide867@gmail.com") #esse é o meu email secundário para testes
pyautogui.press("tab") # seleciona o email
pyautogui.press("tab") # passar para o campo do assunto

# qual o assunto
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")

# qual o corpo do email
texto = f"""
Prezados,

Segue o relatório de vendas de hoje.

Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {qtde_produtos:,}

Qualquer dúvida estou à disposição.
Abs,
Leonardo Kobayashi
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# enviar o email
pyautogui.click(x=1172, y=1028)
```
