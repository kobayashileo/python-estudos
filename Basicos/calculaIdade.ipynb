Esse script calcula a idade da pessoa com base na data de nascimento fornecida pelo usuario (dd/mm/yyyy),

```python
from datetime import datetime

dt_nascimento = input('Digite a sua data de nascimento (dd/mm/yyyy): ')
dt_nascimento = datetime.strptime(dt_nascimento, '%d/%m/%Y')
dt_atual = datetime.now()
mes_atual, dia_atual, mes_nascimento, dia_nascimento = dt_atual.month, dt_atual.day, dt_nascimento.month, dt_nascimento.day

idade = dt_atual.year - dt_nascimento.year

if mes_nascimento > mes_atual:
    idade -= 1
elif mes_nascimento == mes_atual and dia_nascimento > dia_atual:
    idade -= 1

print(idade)

```
